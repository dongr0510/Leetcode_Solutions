```python
def coinChange(coins, amount) -> int:
        res_lst = [[] for _ in range(amount+1)]
        for each_amount in range(amount+1):
            if each_amount%coins[0] == 0:
                res_lst[each_amount].append(each_amount/coins[0])
            else:
                res_lst[each_amount].append(0)
        for i in range(1, len(coins)+1):
            for each_amount in range(amount+1):
                if each_amount-coins[i-1]>=0:
                    res_lst[each_amount][0] = min(res_lst[each_amount][0], res_lst[each_amount-coins[i-1]][0]+1)
#                     print ('coins[i-1]', coins[i-1])
#                     print ('each_amount-coins[i-1]', each_amount-coins[i-1])
#                     print ('amount', each_amount)
#                     print ('res_lst[each_amount][0]', res_lst[each_amount][0])
    
        return res_lst[amount]
```

```python
coinChange([1, 2, 5, 10], 27)
```
