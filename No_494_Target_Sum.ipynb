```python
def findTargetSumWays(nums, S):
        d = {}
        def dfs(cur, i, d):

            if i < len(nums) and (cur, i) not in d: # 搜索周围节点
                d[(cur, i)] = dfs(cur + nums[i], i + 1, d) + dfs(cur - nums[i],i + 1, d)

            return d.get((cur, i), int(cur == S))   
        return dfs(0, 0, d)
```
